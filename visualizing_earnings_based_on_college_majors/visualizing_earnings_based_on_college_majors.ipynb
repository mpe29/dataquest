# Visualizing Earnings Based On College Majors 

In this guided project, we'll explore how using the pandas plotting functionality along with the Jupyter notebook interface allows us to explore data quickly using visualizations.

We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their Github repo.

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

- Rank - Rank by median earnings (the dataset is ordered by this column).
- Major_code - Major code.
- Major - Major description.
- Major_category - Category of major.
- Total - Total number of people with major.
- Sample_size - Sample size (unweighted) of full-time.
- Men - Male graduates.
- Women - Female graduates.
- ShareWomen - Women as share of total.
- Employed - Number employed.
- Median - Median salary of full-time, year-round workers.
- Low_wage_jobs - Number in low-wage service jobs.
- Full_time - Number employed 35 hours or more.
- Part_time - Number employed less than 35 hours.

## Generating Scatter Plots

```python
size_vs_med = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
size_vs_med.set_title('Median vs. Sample_size')
```

```python
size_vs_unem = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
size_vs_unem.set_title('Unemployment_rate vs. Sample_size')
```

```python
full_time_vs_med = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
full_time_vs_med.set_title('Median vs. Full_time')
```

```python
wom_vs_unem = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
wom_vs_unem.set_title('ShareWomen vs. Unemployment_rate')
```

```python
wom_vs_med = recent_grads.plot(x='Women', y='Median', kind='scatter')
wom_vs_med.set_title('Women vs. Median')
```

```python
men_vs_med = recent_grads.plot(x='Men', y='Median', kind='scatter')
men_vs_med.set_title('Men vs. Median')
```

Firstly in must be noted that the plots above are often misleading as they are not normalized. And so it is difficult if a waste of time to compare these different datapoints before normalizing. As we can see, large samples cause the axes to expand and cause the majority of points to condense. 

1. Do students in more popular majors make more money?
We cannot make any cliams on this point with the non-normalized plots and with these comparisons.

2. Do students that majored in subjects that were majority female make more money?
It seems from first glance (No regression attempted yet) that women meadian income is lower than mens.

3. Is there any link between the number of full-time employees and median salary?
We cant see any relationships in this regard with the information above.
